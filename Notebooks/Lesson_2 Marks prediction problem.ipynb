```python
%matplotlib inline
from sklearn.linear_model import LinearRegression
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [15, 10]
```

```python
# SETUP
X, y = make_regression(n_features=1, noise=20,bias=2,random_state=2)
X=X+4
y=(y+200)/4
plt.scatter(X,y)
plt.ylabel('Marks')
plt.xlabel('Number of hours spent studying')
plt.show()
```

```python
lr = LinearRegression()
lr.fit(X,y)
lr.score(X,y)
```

```python
plt.scatter(X,y)
plt.plot(X,lr.predict(X),color='black')
plt.ylabel('Marks')
plt.xlabel('Number of hours spent studying')
plt.show()
```
